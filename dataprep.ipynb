```python
#Series
#The Section is about series in python 
g = pd.Series(range(1,10))
w = pd.Series(["I", "like", "to", "use", "Python", "for", "Data","Science","mostly."],index=range(0,9))
print(g,w)
```

```python
start_date = "Jan 1, 2021"
end_date = "Dec 31, 2021"
idx = pd.date_range(start_date, end_date)
print(f"Number of Days in 2021 is: {len(idx)}")
```

```python
start_date = "Jan 1, 2021"
end_date = "Dec 31, 2021"
idx = pd.date_range(start_date, end_date)
print(f"Number of Days in 2021 is: {len(idx)}")
items_sold = pd.Series(np.random.randint(1,50, size=len(idx)), index=idx)
df=pd.DataFrame([idx, items_sold])
df = df.T.rename(columns = {0:"Date",1:"Sales by Date"})
```
